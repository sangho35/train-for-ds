```python
import pandas_datareader.data as web
import pandas as pd
```

```python
df = web.DataReader('005930', 'naver', start='2009-11-16', end='2020-11-15')
df.head()
```

```python
df = df.astype('int64')
df.info()
```

```python
df.tail()
```

```python
idx = pd.date_range('2009-11-16','2020-11-15')
idx
```

```python
df = df.reindex(idx)
df.tail()
```

```python
df = df.interpolate()
df.tail()
```

```python
df = df.reset_index()
df
```

```python
df = df[['index','Close','Volume']]
df.columns = ['Date','Close','Volume']
df
```
