```python
import pandas as pd
import numpy as np
```

```python
df = pd.read_csv('data/DJIA_data.csv')
df.head()
```

```python
df = df[['Date','Close','Volume']]
df
```

```python
df['Date'] = pd.to_datetime(df['Date'])
df.info()
```

```python
idx = pd.date_range('2006-12-28','2016-12-31')
idx
```

```python
df = df.set_index('Date')
df
```

```python
df = df.reindex(idx)
df
```

```python
df = df.interpolate()
df
```

```python
df = df.reset_index()
```

```python
df.columns=['Date','Close','Volume']
```

```python
df.to_csv('data/DJIA_adj_ver01.csv', index=False)
```
