```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline
```

```python
df = pd.read_csv('data/wine.csv')
df.head()
```

```python
sns.distplot(df['alcohol']);
```

```python
sns.distplot(df['alcohol'], 
             kde=False, 
             bins=10);
```

```python
sns.distplot(df['alcohol'], 
             hist=False,
             rug=True);
```

```python
sns.distplot(df['alcohol'], 
             hist=False,
             rug=True,
             kde_kws={'shade':True});
```

```python
sns.regplot(x='alcohol', y='pH', data=df);
```

```python
sns.regplot(x='alcohol', y='pH', data=df, ci=None);
```

```python
sns.regplot(x='alcohol', y='pH', data=df, fit_reg=False);
```

```python
sns.regplot(x='alcohol', y='quality', data=df, y_jitter=0.2);
```

```python
sns.lmplot(x='alcohol', y='quality', data=df);
```

```python
sns.lmplot(x='alcohol', y='quality', data=df, hue='type');
```

```python
sns.lmplot(x='alcohol', y='quality', data=df, col='type');
```

```python
for style in ['white', 'dark', 'whitegrid', 'darkgrid', 'ticks']:
    sns.set_style(style)
    sns.distplot(df['alcohol']).set_title(style)
    plt.show()
```

```python
sns.distplot(df['alcohol'], color='g');
```

```python
for p in sns.palettes.SEABORN_PALETTES:
    sns.set_palette(p)
    sns.lmplot(x='alcohol', y='quality', data=df, hue='type');
    plt.show()
```

```python
for p in sns.palettes.SEABORN_PALETTES:
    sns.set_palette(p)
    sns.palplot(sns.color_palette())
    print(p)
    plt.show()
```

```python
fig, ax = plt.subplots()
sns.distplot(df['alcohol'], ax=ax)
ax.set(xlabel='Alcohol',
       ylabel='Distribution',
       xlim=(8,14),
       title='Alcohol');
```

```python
fig, (ax0, ax1) = plt.subplots(nrows=1, ncols=2, sharey=True, figsize=(7,4))
sns.distplot(df['alcohol'], ax=ax0)
sns.distplot(df['alcohol'], ax=ax1)
ax0.set(ylabel='Distribution',
        xlabel='Alcohol')
ax1.set(xlabel='Alcohol',
       xlim=(8,14),
       title='Alcohol')
ax1.axvline(x=df['alcohol'].mean(), label='Mean', linestyle='--')
ax1.legend();
```
