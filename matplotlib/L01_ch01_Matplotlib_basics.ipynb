```python
weather = pd.read_csv('data/weather.csv', encoding='euc-kr')
weather.head()
```

```python
olympics = pd.read_csv('data/olympics.csv', index_col='Country')
olympics.head()
```

```python
fig, ax = plt.subplots(figsize=(10,6))
ax.bar(olympics.index, olympics['Gold'])

ax.set_xticklabels(olympics.index, rotation=90)
ax.set_ylabel('금메달 수');
```

```python
fig, ax = plt.subplots()
ax.bar(olympics.index, olympics['Gold'])
ax.bar(olympics.index, olympics['Silver'], bottom=olympics['Gold'])
ax.bar(olympics.index, olympics['Bronze'], bottom=olympics['Gold']+olympics['Silver'])

ax.set_xticklabels(olympics.index, rotation=90)
ax.set_ylabel('메달 수');
```

```python
fig, ax = plt.subplots()
ax.bar(olympics.index, olympics['Gold'], label='금메달')
ax.bar(olympics.index, olympics['Silver'], bottom=olympics['Gold'], label='은메달')
ax.bar(olympics.index, olympics['Bronze'], bottom=olympics['Gold']+olympics['Silver'], label='동메달')

ax.set_xticklabels(olympics.index, rotation=90)
ax.set_ylabel('메달 수')

ax.legend();
```
