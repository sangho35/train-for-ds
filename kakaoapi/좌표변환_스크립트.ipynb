```python
import pandas as pd
import numpy as np
from pyproj import Proj, transform
from pyproj import Transformer
from tqdm import tqdm

import warnings
warnings.filterwarnings('ignore')
```

```python
#WGS 및 TM128 변수 정의

WGS84 = {'proj':'latlong', 'datum':'WGS84', 'ellps':'WGS84'}


TM128 = { 'proj':'tmerc', 'lat_0':'38N', 'lon_0':'128E', 'ellps':'bessel',
   'x_0':'400000', 'y_0':'600000', 'k':'0.9999',
   'towgs84':'-146.43,507.89,681.46'}

trans_data = df.copy()

# xpos_list = []
# ypos_list = []

# for idx, row in tqdm(trans_data.iterrows()):
#     x,y = row['xpos']/10,row['ypos']/10
#     wgs_x, wgs_y = transform(Proj(**TM128),Proj(**WGS84), x, y)
#     xpos_list.append(wgs_x)
#     ypos_list.append(wgs_y)

# trans_data['wgs_xpos'] = xpos_list
# trans_data['wgs_ypos'] = ypos_list


### 연산속도 높이기

converted = transform(Proj(**TM128),Proj(**WGS84), trans_data['xpos'].values/10, trans_data['ypos'].values/10)

trans_data['wgs_xpos'] = converted[0]
trans_data['wgs_ypos'] = converted[1]
trans_data.tail()
```
